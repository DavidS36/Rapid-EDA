### Question 1: How does spotify's API measurements relate to the songs popularity?
For this I decided to create scatter plots that compares various measures that spotify creates against the categories that measure popularity

```python
# Create scatter plots that compares the spotify API measues with stream count and playlists
```

```python
for x in spotify.columns:
    print(x)
```

I first did it like so, but quickly realized that it would be much easier to read the graphs if they were side by side

```python
API_cats = ['bpm' , 'danceability_%', 'valence_%',	'energy_%',	'acousticness_%',	'instrumentalness_%',	'liveness_%',	'speechiness_%']
POP_cats = ['streams', 'in_spotify_playlists', 'in_apple_playlists']
for x1 in API_cats:
  for y1 in POP_cats:
    y = spotify[y1].head(100)
    x = spotify[x1].head(100)
    plt.figure(figsize=(8, 6))
    plt.scatter(x, y, color='blue', marker='o', s=50, alpha=0.7, label='Data Points')
    plt.title(f'{x1} by {y1} Scatter Plot')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.grid(True)

    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
API_cats = ['bpm', 'danceability_%', 'valence_%', 'energy_%', 'acousticness_%', 'instrumentalness_%', 'liveness_%', 'speechiness_%']
POP_cats = ['streams', 'in_spotify_playlists', 'in_apple_playlists']

fig, axs = plt.subplots(len(API_cats), len(POP_cats), figsize=(18, 24))

for i, x1 in enumerate(API_cats):
    for j, y1 in enumerate(POP_cats):
        ax = axs[i, j]
        x = spotify[x1].head(100)
        y = spotify[y1].head(100)

        ax.scatter(x, y, color='blue', marker='o', s=50, alpha=0.7, label='Data Points')
        ax.set_title(f'{x1} by {y1}')
        ax.set_xlabel(x1)
        ax.set_ylabel(y1)
        ax.grid(True)
        ax.legend()

plt.tight_layout()
plt.show()
```

Much better!!!

### Question 2: Which artists where the most popular in 2023
For this, I used the same measures of popularity but then found each artists total and charted them against each other

```python
# Total Amount of streams by artists (barplot)
```

```python
spotify['artist(s)_name'].value_counts().head(10)
```

```python
spotify['streams'] = spotify['streams'].astype(int)
```

```python
artists_streams = spotify.groupby('artist(s)_name')['streams'].sum().sort_values(ascending=False).nlargest(50)
artists_streams
```

```python
fig, ax = plt.subplots(figsize=(8, 10))
sns.barplot(x=artists_streams.values, y=artists_streams.index, ax=ax)
ax.set_yticklabels(artists_streams.index)
ax.set_title('Total Streams by Artist (Top 50)')
ax.set_xlabel('Total Streams (Tens of Billions)')
ax.set_ylabel('Artist')
plt.tight_layout()
plt.show()
```

Cool! It works, lets make it a fucntion, to quickly get the other measures of popularity

```python
def top_50(ranking="streams"):
   x_cat = spotify.groupby('artist(s)_name')[ranking].sum().sort_values(ascending=False).nlargest(50)
   fig, ax = plt.subplots(figsize=(8, 10))
   sns.barplot(x=x_cat.values, y=x_cat.index, ax=ax)
   ax.set_yticklabels(x_cat.index)
   ax.set_title(f'Total "{ranking}" by Artist (Top 50)')
   ax.set_xlabel(f'Total "{ranking}"')
   ax.set_ylabel('Artist')
   plt.tight_layout()
   plt.show()
```

```python
top_50()
```

```python
top_50('in_spotify_playlists')
```

```python
top_50('in_apple_playlists')
```

### Question 3: Which artist had the highest average amount of streams per song?

For this I grouped by artists and found each artists average stream count and then plotted them in a barplot



```python
#Which artist had the highest average amount of streams
```

```python
top_50_avg = spotify.groupby('artist(s)_name')['streams'].mean().sort_values(ascending=False).head(50)
```

```python
fig, ax = plt.subplots(figsize=(8, 10))
sns.barplot(x=top_50_avg.values, y=top_50_avg.index,)
ax.set_yticklabels(top_50_avg.index)
ax.set_title('Average Amount of Streams Per Song by Artist')
ax.set_xlabel('Average Streams')
ax.set_ylabel('Artist')
plt.tight_layout()
plt.show()
```
