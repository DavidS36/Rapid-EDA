#### Here I upload a function I previously used to get a quicklook at my dataset

```python
def quicklook(df):
    print("First 2 rows of the DataFrame:")
    display(df.head(2))

    print("\nDescriptive statistics of numerical columns:")
    display(df.describe())

    print("\nDescriptive statistics of object columns:")
    display(df.describe(include='O'))

    print("\nInformation about the DataFrame:")
    df.info()

    print("\nShape of the DataFrame:")
    print(df.shape)

    print("\nNumber of missing values in each column:")
    print(df.isnull().sum())

    print("\nNumber of duplicated rows:")
    print(df.duplicated().sum())

    print("\nNumber of unique values in each column:")
    print(df.nunique())

    print("\nData types of each column:")
    print(df.dtypes)

    print("\nUnique values in each column:")
    print(df.apply(lambda x: x.unique()))

    print("\nColumn names of the DataFrame:")
    print(df.columns)

    print("\nIndex of the DataFrame:")
    print(df.index)

    print("\nRandom sample of 5 rows from the DataFrame:")
    display(df.sample(5))

    print("\nValue counts of each unique value in the DataFrame:")
    display(df.apply(lambda x: x.value_counts()))
```

```python
def highest_salary(filter):
  dds_high_sal = dds.groupby(filter)['salary_in_usd'].mean().sort_values(ascending=False).head(10)
  fig, ax = plt.subplots(figsize=(10, 6))
  sns.barplot(x=dds_high_sal.values, y=dds_high_sal.index, ax=ax)
  ax.set_title(f'Highest Salary by {filter}')
  ax.set_xlabel('Salary in USD')
  ax.set_ylabel(filter)
  plt.show()
```

```python
#based on experience level what job title has the highest average salary?
```

```python
for exp in dds['experience_level'].unique():
  set_exp = dds[dds['experience_level'] == exp]
  set_exp_sal = set_exp.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False).head(10)
  fig, ax = plt.subplots(figsize=(10, 6))
  sns.barplot(x=set_exp_sal.values, y=set_exp_sal.index, ax=ax)
  ax.set_title(f'{exp} highest average salary by Job Title')
  ax.set_xlabel('Salary in USD')
  ax.set_ylabel('Job Title')
  plt.show()
```

```python
experience_levels = dds['experience_level'].unique()
fig, axes = plt.subplots(2, 2, figsize=(20, 12))
axes = axes.flatten()
for i, exp in enumerate(experience_levels):
    set_exp = dds[dds['experience_level'] == exp]
    set_exp_sal = set_exp.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False).head(10)
    sns.barplot(x=set_exp_sal.values, y=set_exp_sal.index, ax=axes[i])
    axes[i].set_title(f'{exp} highest average salary by Job Title')
    axes[i].set_xlabel('Salary in USD')
    axes[i].set_ylabel('Job Title')
plt.tight_layout()
plt.show()
```
